# Norm+PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Prepare data
# Exclude target variables ("vomitoxin_ppb", "log_vomitoxin_ppb") and identifier
# ("hsi_id")
spectral_cols = [c for c in df.columns if c not in ["vomitoxin_ppb", "log_vomitoxin_ppb", "hsi_id"]]

# Define feature matrix (X) and target variables (y_raw, y_log)
X = df[spectral_cols]
y_raw = df["vomitoxin_ppb"]
y_log = df["log_vomitoxin_ppb"]

# Train-Test Split
X_train, X_test, y_train_raw, y_test_raw, y_train_log, y_test_log = train_test_split(
    X, y_raw, y_log, test_size=0.2, random_state=42
)

# Standardize the training data and apply the same transformation to test data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Determine maximum possible components
max_components = min(X_train_scaled.shape[0], X_train_scaled.shape[1])

# PCA with all possible components for scree plot (up to max_components)
pca_full = PCA(n_components=max_components, random_state=42)
pca_full.fit(X_train_scaled)

# Compute cumulative explained variance
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)

# Plot the scree plot to analyze variance retention
plt.plot(cumulative_variance, marker='o')
plt.axhline(y=0.965, color='r', linestyle='--')
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")
plt.show()

# Find the smallest number of components needed for ~97% variance
candidate_index = np.argmax(cumulative_variance >= 0.965) + 1
n_components = min(candidate_index, max_components)  # ensure it doesn't exceed max_components
print(f"Components for ~97% variance (bounded by {max_components}): {n_components}")

# Apply PCA transformation with the selected number of components
pca = PCA(n_components=n_components, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Visualize first two components
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train_log, cmap='viridis', alpha=0.7)
plt.colorbar(label='Log Vomitoxin (PPB)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title(f'PCA (2D) - {n_components} Components Retained')
plt.show()

# Train a linear regression model using the PCA-transformed features (Raw Target)
reg_raw = LinearRegression()
reg_raw.fit(X_train_pca, y_train_raw)
y_pred_raw = reg_raw.predict(X_test_pca)

# Calculate RMSE and R² score for raw target predictions
rmse_raw = np.sqrt(mean_squared_error(y_test_raw, y_pred_raw))
r2_raw = r2_score(y_test_raw, y_pred_raw)

# Linear Regression (Log Target) but compare on original scale
reg_log = LinearRegression()
reg_log.fit(X_train_pca, y_train_log)
y_pred_log = reg_log.predict(X_test_pca)
y_pred_log_original = np.exp(y_pred_log) - 1
rmse_log_original = np.sqrt(mean_squared_error(y_test_raw, y_pred_log_original))
r2_log_original = r2_score(y_test_raw, y_pred_log_original)

print("---- Raw Target ----")
print(f"RMSE: {rmse_raw:.3f}")
print(f"R^2 : {r2_raw:.3f}")

print("\n---- Log Target (original scale) ----")
print(f"RMSE: {rmse_log_original:.3f}")
print(f"R^2 : {r2_log_original:.3f}")
```

# Function for mse,mae,log

```python
def inverse_log_preds(log_preds):
    """Convert log(y+1) predictions back to original scale."""
    return np.exp(log_preds) - 1

def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2

def print_scores(label, mae, rmse, r2):
    print(f"{label}: MAE={mae:.3f}, RMSE={rmse:.3f}, R²={r2:.3f}")
```

# **DEEP CNN**

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def build_deep_cnn(input_shape):
    model = keras.Sequential([
        layers.Conv1D(32, kernel_size=3, activation='relu', input_shape=input_shape),
        layers.Conv1D(32, kernel_size=3, activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),

        layers.Conv1D(64, kernel_size=3, activation='relu'),
        layers.Conv1D(64, kernel_size=3, activation='relu'),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling1D(),

        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(1)  # for regression
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

def train_deep_cnn(X_train, X_test, y_train_raw, y_test_raw, y_train_log, y_test_log, label, epochs=20):
    # Reshape for Conv1D: (samples, length, 1)
    X_train_cnn = np.expand_dims(X_train, axis=-1)
    X_test_cnn = np.expand_dims(X_test, axis=-1)
    input_shape = X_train_cnn.shape[1:]

    # --- Raw target
    model_raw = build_deep_cnn(input_shape)
    model_raw.fit(X_train_cnn, y_train_raw, epochs=epochs, batch_size=32, verbose=0)
    preds_raw = model_raw.predict(X_test_cnn).flatten()
    mae_raw, rmse_raw, r2_raw = regression_metrics(y_test_raw, preds_raw)

    # --- Log target (evaluate in original scale)
    model_log = build_deep_cnn(input_shape)
    model_log.fit(X_train_cnn, y_train_log, epochs=epochs, batch_size=32, verbose=0)
    preds_log = model_log.predict(X_test_cnn).flatten()
    preds_log_orig = inverse_log_preds(preds_log)
    mae_log, rmse_log, r2_log = regression_metrics(y_test_raw, preds_log_orig)

    print(f"=== Deeper CNN on {label} Data ===")
    print_scores("Raw Target    ", mae_raw, rmse_raw, r2_raw)
    print_scores("Log (original)", mae_log, rmse_log, r2_log)
    print()

##############################################################################
# CNN on NON-REDUCED data
train_deep_cnn(
    X_train_nr_sc, X_test_nr_sc, y_train_raw_nr, y_test_raw_nr,
    y_train_log_nr, y_test_log_nr, label="Non-Reduced", epochs=20
)

# CNN on PCA-REDUCED data
train_deep_cnn(
    X_train_r_sc, X_test_r_sc, y_train_raw_r, y_test_raw_r,
    y_train_log_r, y_test_log_r, label="PCA-Reduced", epochs=20
)
```

# **DEEP LSTM**

```python
def build_deep_lstm(input_shape):
    model = keras.Sequential([
        layers.LSTM(32, return_sequences=True, input_shape=input_shape),
        layers.LSTM(32),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

def train_deep_lstm(X_train, X_test, y_train_raw, y_test_raw, y_train_log, y_test_log, label, epochs=20):
    # Reshape for LSTM: (samples, timesteps, features) => (samples, #bands, 1)
    X_train_lstm = np.expand_dims(X_train, axis=-1)
    X_test_lstm = np.expand_dims(X_test, axis=-1)
    input_shape = X_train_lstm.shape[1:]  # (#bands, 1)

    # --- Raw target
    model_raw = build_deep_lstm(input_shape)
    model_raw.fit(X_train_lstm, y_train_raw, epochs=epochs, batch_size=32, verbose=0)
    preds_raw = model_raw.predict(X_test_lstm).flatten()
    mae_raw, rmse_raw, r2_raw = regression_metrics(y_test_raw, preds_raw)

    # --- Log target
    model_log = build_deep_lstm(input_shape)
    model_log.fit(X_train_lstm, y_train_log, epochs=epochs, batch_size=32, verbose=0)
    preds_log = model_log.predict(X_test_lstm).flatten()
    preds_log_orig = inverse_log_preds(preds_log)
    mae_log, rmse_log, r2_log = regression_metrics(y_test_raw, preds_log_orig)

    print(f"=== Deeper LSTM on {label} Data ===")
    print_scores("Raw Target    ", mae_raw, rmse_raw, r2_raw)
    print_scores("Log (original)", mae_log, rmse_log, r2_log)
    print()

##############################################################################
# LSTM on NON-REDUCED data
train_deep_lstm(
    X_train_nr_sc, X_test_nr_sc, y_train_raw_nr, y_test_raw_nr,
    y_train_log_nr, y_test_log_nr, label="Non-Reduced", epochs=20
)

# LSTM on PCA-REDUCED data
train_deep_lstm(
    X_train_r_sc, X_test_r_sc, y_train_raw_r, y_test_raw_r,
    y_train_log_r, y_test_log_r, label="PCA-Reduced", epochs=20
)
```

# TRANSFORMER

```python
def build_transformer_model(input_shape):
    """
    Creates a simple Transformer-based regression model:
      - input_shape: (sequence_length, 1)
    """
    inputs = keras.Input(shape=input_shape)

    # 1) Project 1-D input (band reflectance) to a higher dimension (e.g., 32)
    x = layers.Dense(32)(inputs)

    # 2) Self-Attention Block
    attn_output = layers.MultiHeadAttention(num_heads=2, key_dim=32)(x, x)
    x = layers.Add()([x, attn_output])
    x = layers.LayerNormalization()(x)

    # 3) Feed Forward Block
    ff = layers.Dense(64, activation='relu')(x)
    ff = layers.Dense(32)(ff)
    x = layers.Add()([x, ff])
    x = layers.LayerNormalization()(x)

    # 4) Global Pooling and Output
    x = layers.GlobalAveragePooling1D()(x)       # [batch_size, hidden_dim]
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(1)(x)                 # Regression output

    model = keras.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
def train_transformer(
    X_train, X_test,
    y_train_raw, y_test_raw,
    y_train_log, y_test_log,
    label, epochs=20
):
    """
    1) Reshape input for the Transformer: (samples, timesteps, features).
    2) Train two models: raw target + log target (inverted for final metrics).
    3) Print results.
    """
    # (samples, #bands, 1)
    X_train_seq = np.expand_dims(X_train, axis=-1)
    X_test_seq = np.expand_dims(X_test, axis=-1)
    input_shape = X_train_seq.shape[1:]  # (#bands, 1)

    # -----------------------------
    # (A) Transformer on RAW target
    # -----------------------------
    model_raw = build_transformer_model(input_shape)
    model_raw.fit(
        X_train_seq, y_train_raw,
        epochs=epochs, batch_size=32, verbose=0
    )
    preds_raw = model_raw.predict(X_test_seq).flatten()
    mae_raw, rmse_raw, r2_raw = regression_metrics(y_test_raw, preds_raw)

    # -----------------------------
    # (B) Transformer on LOG target
    # -----------------------------
    model_log = build_transformer_model(input_shape)
    model_log.fit(
        X_train_seq, y_train_log,
        epochs=epochs, batch_size=32, verbose=0
    )
    preds_log = model_log.predict(X_test_seq).flatten()
    preds_log_orig = inverse_log_preds(preds_log)
    mae_log, rmse_log, r2_log = regression_metrics(y_test_raw, preds_log_orig)

    # -----------------------------
    # Reporting
    # -----------------------------
    print(f"=== Transformer on {label} Data ===")
    print_scores("Raw Target    ", mae_raw, rmse_raw, r2_raw)
    print_scores("Log (original)", mae_log, rmse_log, r2_log)
    print()
```

```python

# Transformer on NON-REDUCED Data
train_transformer(
    X_train_nr_sc, X_test_nr_sc,
    y_train_raw_nr, y_test_raw_nr,
    y_train_log_nr, y_test_log_nr,
    label="Non-Reduced",
    epochs=20
)

# Transformer on PCA-REDUCED Data
train_transformer(
    X_train_r_sc, X_test_r_sc,
    y_train_raw_r, y_test_raw_r,
    y_train_log_r, y_test_log_r,
    label="PCA-Reduced",
    epochs=20
)
```

# **Random forest With hyperparameter Tuning**

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# ---------------------------
# 1. DATA PREPARATION
# ---------------------------
spectral_cols = [c for c in df.columns if c not in ["vomitoxin_ppb", "log_vomitoxin_ppb", "hsi_id"]]
X = df[spectral_cols]
y_raw = df["vomitoxin_ppb"]
y_log = df["log_vomitoxin_ppb"]

X_train, X_test, y_train_raw, y_test_raw, y_train_log, y_test_log = train_test_split(
    X, y_raw, y_log, test_size=0.2, random_state=42
)

# ---------------------------
# 2. SCALING & PCA
# ---------------------------
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

max_components = min(X_train_scaled.shape[0], X_train_scaled.shape[1])
pca_full = PCA(n_components=max_components, random_state=42)
pca_full.fit(X_train_scaled)
cumulative_var = np.cumsum(pca_full.explained_variance_ratio_)

# Choose ~96.5% variance for example
n_components = np.argmax(cumulative_var >= 0.965) + 1
n_components = min(n_components, max_components)

pca = PCA(n_components=n_components, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# ---------------------------
# 3. RANDOM FOREST + HYPERPARAM TUNING
# ---------------------------
param_grid = {
    'n_estimators': [20, 30, 50, 100, 150, 200],
    'max_depth': [None, 5, 10, 15, 20, 30],
    'min_samples_split': [2, 5, 10, 15]
}

# (A) Random Forest for RAW Target
rf_raw = RandomForestRegressor(random_state=42)
grid_raw = GridSearchCV(rf_raw, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_raw.fit(X_train_pca, y_train_raw)

best_raw = grid_raw.best_estimator_
pred_raw = best_raw.predict(X_test_pca)
mae_raw = mean_absolute_error(y_test_raw, pred_raw)
rmse_raw = np.sqrt(mean_squared_error(y_test_raw, pred_raw))
r2_raw = r2_score(y_test_raw, pred_raw)

# (B) Random Forest for LOG Target (evaluation on log scale & original scale)
rf_log = RandomForestRegressor(random_state=42)
grid_log = GridSearchCV(rf_log, param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_log.fit(X_train_pca, y_train_log)

best_log = grid_log.best_estimator_
pred_log = best_log.predict(X_test_pca)  # This is in log scale
mae_log = mean_absolute_error(y_test_raw, np.exp(pred_log) - 1)
rmse_log = np.sqrt(mean_squared_error(y_test_raw, np.exp(pred_log) - 1))
r2_log = r2_score(y_test_raw, np.exp(pred_log) - 1)
```

```python
# ---------------------------
# 4. METRICS (Raw vs. Log)
# ---------------------------
print("=== RandomForest (Raw Target) ===")
print("Best Params:", grid_raw.best_params_)
print(f"MAE={mae_raw:.3f}, RMSE={rmse_raw:.3f}, R²={r2_raw:.3f}\n")

print("=== RandomForest (Log Target) - Evaluated on Log Scale ===")
print("Best Params:", grid_log.best_params_)
print(f"MAE={mae_log:.3f}, RMSE={rmse_log:.3f}, R²={r2_log:.3f}\n")

# ---------------------------
# 5. VISUALIZATION:
#    Scatter Plot of Log Actual vs. Log Predicted
# ---------------------------
plt.scatter(y_test_log, np.log(pred_raw + 1), alpha=0.6)
plt.xlabel("Log Actual (log_vomitoxin_ppb)")
plt.ylabel("Log Predicted (Model Output)")
plt.title("RandomForest - HyperTuned (Raw_Target) | Log Actual vs. Log Predicted")
plt.show()

plt.scatter(y_test_log, pred_log, alpha=0.6)
plt.xlabel("Log Actual (log_vomitoxin_ppb)")
plt.ylabel("Log Predicted (Model Output)")
plt.title("RandomForest - HyperTuned (Log_Target) | Log Actual vs. Log Predicted")
plt.show()

# ---------------------------
# 6. SUMMARY
# ---------------------------
print("Summary of Model Performance and Limitations:")
print("1) Metrics (Log Scale):")
print("   • MAE, RMSE, and R² indicate how well the model fits the log-transformed target.")
print("   • If the data is highly skewed, log-transform can help stabilize variance.")
print("2) The scatter plot shows how close log predictions are to the actual log values.")
print("3) Possible Limitations:")
print("   • PCA may discard some relevant variance if set too aggressively.")
print("   • More exhaustive hyperparameter tuning (e.g., more parameters, random/grid search) could yield better results.")
print("   • Real-world sensor noise and data variability can impact performance.")
```
